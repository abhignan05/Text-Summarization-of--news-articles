```python
from transformers import pipeline
from transformers import PegasusForConditionalGeneration, PegasusTokenizer
```

```python
model = "google/pegasus-xsum"
pegasus_tokentizer = PegasusTokenizer.from_pretrained(model)
```

```python
example_text = input("Enter the text you want to summarize: ")
```

```python
summarizer = pipeline(
    "summarization",
    model=model,
    tokenizer=pegasus_tokentizer,
    framework="pt"
)
```

```python
summary = summarizer(example_text, min_length=30, max_length= 60)
summary[0]["summary_text"]
```
